```python
# import numpy as np
import numpy as np
import ipyvolume as ipv

sideLength=15
stepSize=1/sideLength

oS=np.zeros((sideLength,sideLength,sideLength))

oS

xi=0
while xi<oS.shape[0]:
    yi=0
    while yi<oS.shape[1]:
        zi=0
        while zi<oS.shape[2]:
    #        print(xi)
    #        print(np.sin(np.pi/2))
           
            oS[xi,yi,zi]=np.abs(np.sin(np.pi/2*(xi*stepSize)*(yi*stepSize)*(zi*stepSize)))
    #        print(np.sin(np.pi/2*stepSize*xi*yi*zi))
    #        print(oS[xi,yi,zi])
            zi+=1
        
        yi+=1
        
    
    xi+=1
```

```python



ipv.figure()
ipv.volshow(oS, level=[0.25, 0.75], opacity=0.03, level_width=0.1, data_min=0, data_max=1)
ipv.view(-30, 40)
ipv.show()
```

```python
import scipy.fft as fft

fS=fft.fftshift( fft.fft(oS))
print(fS)
```

```python
import ipyvolume as ipv2

ipv2.figure()
ipv2.volshow(fS, level=[0.25, 0.75], opacity=0.03, level_width=0.1, data_min=0, data_max=1)
ipv2.view(-30, 40)
ipv2.show()
```
